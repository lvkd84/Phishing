```python
from scipy.io import arff
import pandas as pd

#load_data = arff.loadarff('Training_Dataset.arff')
load_data = arff.loadarff('PhishingData.arff')
data = pd.DataFrame(load_data[0])
data
```

```python
from pgmpy.estimators import HillClimbSearch, BicScore
est = HillClimbSearch(data, scoring_method=BicScore(data))
best_model = est.estimate()
sorted(best_model.nodes())
best_model.edges()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
nx.draw(best_model, with_labels=True)
plt.show()
```

```python
from pgmpy.estimators import ConstraintBasedEstimator
from pgmpy.models import BayesianModel
est_PC = ConstraintBasedEstimator(data)
model_PC = est_PC.estimate()
sorted(model_PC.nodes())
model_PC.edges()
```
